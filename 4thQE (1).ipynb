```python
import numpy as np
import matplotlib.pyplot as plt

print("Problem 1: Euler Method")
print("Assumed differential equation: dy/dx = x + y")

# Define the differential equation f(x, y)
def f_euler(x, y):
    return x + y

def euler_method_simplified(f, x0, y0, h, xn):
    x_values = [x0]
    y_values = [y0]
    
    x = x0
    y = y0
    
    while x < xn:
        y_new = y + h * f(x, y)
        x_new = x + h
        
        x_values.append(x_new)
        y_values.append(y_new)
        
        x = x_new
        y = y_new
        
    return x_values, y_values

# Problem parameters
x0_p1 = 0.0
y0_p1 = 1.0
h_p1 = 0.5
xn_p1 = 4.0

# Solve using Euler method
x_euler, y_euler = euler_method_simplified(f_euler, x0_p1, y0_p1, h_p1, xn_p1)

print("\nFinal Results (Euler Method)")
for i in range(len(x_euler)):
    print(f"x = {x_euler[i]:.2f}, y = {y_euler[i]:.4f}")

# Plotting the results
plt.figure(figsize=(8, 6))
plt.plot(x_euler, y_euler, marker='o', linestyle='-', color='b', label='Euler Method')
plt.title('Problem 1: Euler Method Numerical Integration')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

#the graph shows a step-by-step numerical approximation of an exponentially growing function, where each step is a straight line segment, and due to the method's characteristics and the function's concavity, it tends to fall below the true solution.
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("Problem 2: Heun's Method")
print("Assumed differential equation: dy/dx = x + y")

# Define the differential equation f(x, y)
def f_heun(x, y):
    return x + y

def heuns_method_simplified(f, x0, y0, h, x_target):
    x_values = [x0]
    y_values = [y0]
    
    x = x0
    y = y0
    
    while round(x, 10) < round(x_target, 10): 
        # Predictor step (Euler)
        y_predictor = y + h * f(x, y)
        
        # Corrector step
        y_new = y + (h / 2) * (f(x, y) + f(x + h, y_predictor))
        x_new = x + h
        
        x_values.append(x_new)
        y_values.append(y_new)
        
        x = x_new
        y = y_new
        
    return x_values, y_values

# Problem parameters
x0_p2 = 0.0
y0_p2 = 2.0
h_p2 = 0.2
x_target_p2 = 2.0 

# Solve using Heun's method
x_heuns, y_heuns = heuns_method_simplified(f_heun, x0_p2, y0_p2, h_p2, x_target_p2)

print("Final Results (Heun's Method)")
for i in range(len(x_heuns)):
    print(f"x = {x_heuns[i]:.2f}, y = {y_heuns[i]:.4f}")

print(f"\nApproximation for y({x_target_p2:.1f}) is: {y_heuns[-1]:.4f}")

# Plotting the results
plt.figure(figsize=(8, 6))
plt.plot(x_heuns, y_heuns, marker='o', linestyle='-', color='g', label='Heun\'s Method')
plt.title('Problem 2: Heun\'s Method Numerical Integration')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

 #Similar to Problem 1, the differential equation dy/dx=x+y causes the solution to grow exponentially. As x and y increase, the slope x+y increases, leading to a steeper curve. The graph correctly captures this accelerating upward trend.
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("Problem 3: 4th Order RK Method ")
print("Assumed second-order differential equation: y'' = -y (Simple Harmonic Motion)")

# Define the system of first-order differential equations
# Here, y_val is 'u' and y_prime_val is 'v'
def f_rk4_system(x, y_val, y_prime_val):
    """
    Returns the derivatives [u', v'] for the system.
    u' = y' = v
    v' = y'' = f(x, y, y')
    """
    du_dx = y_prime_val  # This is u' = v
    dv_dx = -y_val       # This is v' = -y (from y'' = -y)
    return np.array([du_dx, dv_dx])

def rk4_second_order_simplified(f_sys, x0, y0, y_prime0, h, x_end):
    x_values = [x0]
    y_values = [y0]
    y_prime_values = [y_prime0]
    
    x = x0
    y = y0        # Represents u
    y_prime = y_prime0 # Represents v
    
    while round(x, 10) < round(x_end, 10):
        # Calculate k1 for u and v
        k1 = f_sys(x, y, y_prime)
        k1_y = h * k1[0]
        k1_y_prime = h * k1[1]

        # Calculate k2
        k2 = f_sys(x + 0.5 * h, y + 0.5 * k1_y, y_prime + 0.5 * k1_y_prime)
        k2_y = h * k2[0]
        k2_y_prime = h * k2[1]

        # Calculate k3
        k3 = f_sys(x + 0.5 * h, y + 0.5 * k2_y, y_prime + 0.5 * k2_y_prime)
        k3_y = h * k3[0]
        k3_y_prime = h * k3[1]

        # Calculate k4
        k4 = f_sys(x + h, y + k3_y, y_prime + k3_y_prime)
        k4_y = h * k4[0]
        k4_y_prime = h * k4[1]

        # Update y and y_prime
        y_new = y + (k1_y + 2 * k2_y + 2 * k3_y + k4_y) / 6
        y_prime_new = y_prime + (k1_y_prime + 2 * k2_y_prime + 2 * k3_y_prime + k4_y_prime) / 6
        x_new = x + h
        
        x_values.append(x_new)
        y_values.append(y_new)
        y_prime_values.append(y_prime_new)
        
        x = x_new
        y = y_new
        y_prime = y_prime_new
        
    return x_values, y_values, y_prime_values

# Problem parameters
x0_p3 = 0.0
y0_p3 = 4.0
y_prime0_p3 = 0.0
h_p3 = 0.2
x_end_p3 = 5.0

# Solve using RK4 method
x_rk4, y_rk4, y_prime_rk4 = rk4_second_order_simplified(f_rk4_system, x0_p3, y0_p3, y_prime0_p3, h_p3, x_end_p3)

print("Final Results (4th Order RK Method)")
for i in range(len(x_rk4)):
    print(f"x = {x_rk4[i]:.2f}, y = {y_rk4[i]:.4f}, y' = {y_prime_rk4[i]:.4f}")

# Plotting results for Problem 3
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
plt.plot(x_rk4, y_rk4, marker='o', linestyle='-', color='b', label='y(x)')
plt.title('Problem 3: RK4 - y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
plt.plot(x_rk4, y_prime_rk4, marker='x', linestyle='--', color='r', label="y'(x)")
plt.title("Problem 3: RK4 - y' vs x")
plt.xlabel('x')
plt.ylabel("y'")
plt.grid(True)
plt.legend()

plt.tight_layout() # Adjusts plot parameters for a tight layout
plt.show()

#The output perfectly illustrates the high accuracy of the RK4 method in numerically solving a second-order oscillatory differential equation. The calculated values and the plots beautifully mimic the sinusoidal behavior of the analytical solution for simple harmonic motion.
```
